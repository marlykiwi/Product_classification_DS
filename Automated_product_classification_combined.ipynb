#### Using vectors from word2vec combined with probabilites with images

```python
df_prob = pd.read_csv('df_prob.csv', index_col=0)
```

```python
df_prob.head()
```

```python
feats_prob = df_prob.drop('prdtypecode', axis = 1)
target_prob = df_prob['prdtypecode']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(feats_prob, target_prob, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import model_selection

knn = KNeighborsClassifier(n_neighbors = 3, metric = 'minkowski')
knn.fit(X_train, Y_train)
```

```python
y_pred_prob = knn.predict(X_test)
print("Train Score:", knn.score(X_train, Y_train))
print("Test Score:", knn.score(X_test, Y_test))
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test, y_pred_prob))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
plt.figure(figsize =(30,30))
cm_display = ConfusionMatrixDisplay.from_predictions(Y_test, y_pred_prob, 
                                                     xticks_rotation='vertical', cmap = plt.cm.Blues)
fig = cm_display.ax_.get_figure() 
fig.set_figwidth(20)
fig.set_figheight(20)  
```
